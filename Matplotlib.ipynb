```python
import numpy as np
```

```python
#simple example
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph is 2d')
plt.savefig('test.png')
```

```python
y=x*x
#plt plot
plt.plot(x,y,c='b')
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,y,'b--*')
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('Xaxis')
plt.ylabel('Ylabel')
plt.title('2d Diagram')
```

```python
#creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'bo--')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
#compute the x and y coordinates for the points on a sine curve/
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(0,10,0.1)
y=np.cos(x)
plt.plot(x,y)
```

```python
x=np.arange(0,np.pi*4,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine wave')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine wave')
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
```

## Pie Chart

```python
# Data to Plot
labels='python','c++','Ruby','Java'
sizes=[81,50,65,31]
colors=['gold','green','blue','red']
explode=(0.2,0,0,0)
```

```python
#plot pie chart
plt.pie(sizes,explode=explode,colors=colors,labels=labels,labeldistance=1.2,frame=True,autopct='%1.1f%%',shadow=True,wedgeprops = {'linewidth': 1},radius=1)
```
