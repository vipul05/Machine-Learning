## Seaborn

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

## Correlation with Heatmap

```python
df.corr(method='spearman')
```

```python
sns.heatmap(df.corr())
```

## Univariate Analysis

```python
sns.jointplot(data=df,x='tip',y='total_bill',kind='hex')
```

```python
sns.jointplot(data=df,x='tip',y='total_bill',kind='reg')
```

## PairPlot

#### PairPlot is also a scatterplot, all features are plotted in interrelated way, though it only works for int or float value and not categories

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.displot(df['tip'])
```

```python
sns.histplot(df['tip'],kde=True,bins=20)
```


## Categorical Plots

* Boxplot
* Count plot
* violinplot
* barplot


```python
#count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
#barplot needs both x and y value
sns.barplot(x='sex',y='total_bill',data=df)
```
